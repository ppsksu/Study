```python
import datetime
import numpy as np
import cv2
```

```python
X_LEN = 512
Y_LEN = 672
X_DIFF = 60
Y_DIFF = 80
```

```python
def can_boolean(frame):
    if frame is None:
        return 0
    img = cv2.medianBlur(frame, 5)
    circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 200, param1=50, param2=30, minRadius=220, maxRadius=235)

    if circles is None:
        return 0
    else:
        for i in circles[0, :]:
            ##cv2.circle(frame,(i[0],i[1]),int(i[2]),(0,255,0),7) # 201229_1 반지름은 정수여야 함
            bool_x = (((X_LEN / 2) - X_DIFF) <= i[0]) & (i[0] <= ((X_LEN / 2) + X_DIFF))  # +-60
            bool_y = (((Y_LEN / 2) - Y_DIFF) <= i[1]) & (i[1] <= ((Y_LEN / 2) + Y_DIFF))  # +-80

            if bool_x & bool_y:
                # return '캔O'
                return 1
            else:
                # return '좌표범위X'
                return 0
```

```python
cap = cv2.VideoCapture("D:/ksu/Study/CAN_SMAPLE2.mp4")
success, image = cap.read()

count = 0
success = True
pass_frame = 0

while success:
    success, image = cap.read()
    if image is None:
        continue
    # image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # 그레이로 형식으로 변경
    # image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE) # 시계방향 회전
    image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)  # 반시계방향 회전
    count += 1
    if can_boolean(image) == 1:
        print(str(count) + ' +++ 캔')
        cv2.imwrite("D:/ksu/Study/can_capture/frame%d.jpg" % count, image)
```
